##Getting and cleaning data

```python
import sys
sys.path.append('../../krad/')
import clean, solution
from score import kaggle_metric, heaviside
from functools import partial


filename = "../Data/train_2013.csv"

use_full_data = False

if use_full_data:
    # the whole enchilada
    data = pd.read_csv(filename)
    nrows = len(data)
else:
    # smaller test set for speed
    nrows = 10000
    data = pd.read_csv(filename, nrows=nrows) 
```

```python
# Hack to split multiple radar data into multiple rows
list_df = []

# get a list of all the columns that will have to be split by radar
columns_to_split = list(data.columns)
columns_to_split.remove('Id')
columns_to_split.remove('Expected')

# create an array of all the column names, in the same order as in the 
# original data
right_column_order = ['unique_Id']
for col in data.columns:
    right_column_order.append(col)
 
# append a new column that will represent how many radars were in the set from
# which a given row came from
right_column_order.append('number_of_radars')

id_counter = -1
for index in range(len(data)):    
    row = data.loc[index].copy()

    ID = row['Id']
    expected = row['Expected']

    row['RadarCounts'] = clean.getRadarLength(row['TimeToEnd'])

    number_of_radars = len(row['RadarCounts'])
    list_newrows_dict = []
    for i in row['RadarCounts']:
        id_counter += 1
        list_newrows_dict.append({'unique_Id':id_counter,'Id':ID,
                                  'number_of_radars':number_of_radars,
                                  'Expected':expected})

    for col in columns_to_split:
    
        subrow = row[['RadarCounts',col]]

        for array,dict in zip(clean.separate_listInColumn(subrow),list_newrows_dict):
            dict[col] = N.array(array)

    list_df.append(pd.DataFrame(list_newrows_dict))

clean_df = pd.concat(list_df)[right_column_order].set_index('unique_Id')
del data
```

```python
normalized_expected = clean_df['Expected'].apply(lambda v: N.where( v> 70, 71, v)) 

number_of_gauges    = float(normalized_expected.shape[0])
number_unreasonable = float(sum( normalized_expected > 70))

fraction_of_unreasonable = number_unreasonable / number_of_gauges

normalized_expected.hist(range=(0.,71),bins=71,log=True)   
plt.title('Unreasonable gauge measurements: %4.2f percent'%(100*fraction_of_unreasonable))
plt.ylabel('number of gauges')
plt.xlabel('mm of rain')
plt.show()
```

```python
# extract the data which is reasonable
is_reasonable_data = clean_df['Expected'] < 70

# use the .copy method to avoid those annoying SettingWithCopyWarning 
df = clean_df[is_reasonable_data].copy()

# split the data in two: whether it rained or not.
rain    = df['Expected'].apply(lambda n: n > 0 ).values


def averager(x):
    """
    Average each column, and throw out the NaN
    """
    avg = []
    for i in range(len(x)):
        value = N.mean(x.iloc[i])
        if not N.isfinite(value):
            value = -9999.
        avg.append(value)
    return pd.Series(avg)


from sklearn.decomposition import PCA

error_codes = N.array([-99000.,-99901.,-99903.,999.0,N.NaN])


column_names = ['DistanceToRadar', 
                'Composite', 
                'HybridScan', 
                'HydrometeorType',  
                'RadarQualityIndex', 
                'Reflectivity',
                'ReflectivityQC',  
                'Velocity' ]
                
column_names = ['Composite', 
                'HybridScan', 
                'ReflectivityQC',  
                'Velocity' ]

X = df[column_names].apply(averager,axis=1).values
y = df['Expected'].values

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

for i,v in enumerate(pca.components_):
    print '\n\n Vector %i'%i
    for name, x in zip(column_names,v):
        print '%20s  : %+8.4e'%(name,x)
```

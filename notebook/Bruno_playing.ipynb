##Getting and cleaning data

```python
import sys
sys.path.append('../../krad/')
import clean, solution
from score import kaggle_metric, heaviside
from functools import partial


filename = "../Data/train_2013.csv"

use_full_data = False

if use_full_data:
    # the whole enchilada
    data = pd.read_csv(filename)
    nrows = len(data)
else:
    # smaller test set for speed
    nrows = 1000
    data = pd.read_csv(filename, nrows=nrows)


# Hack to split multiple radar data into multiple rows
list_df = []

columns_to_split = list(data.columns)
columns_to_split.remove('Id')
columns_to_split.remove('Expected')

right_column_order = ['unique_Id']
for col in data.columns:
    right_column_order.append(col)


id_counter = -1
for index in range(len(data)):    
    row = data.loc[index].copy()

    ID = row['Id']
    expected = row['Expected']

    row['RadarCounts'] = clean.getRadarLength(row['TimeToEnd'])

    list_newrows_dict = []
    for i in row['RadarCounts']:
        id_counter += 1
        list_newrows_dict.append({'unique_Id':id_counter,'Id':ID,'Expected':expected})

    for col in columns_to_split:
    
        subrow = row[['RadarCounts',col]]

        for array,dict in zip(clean.separate_listInColumn(subrow),list_newrows_dict):
            dict[col] = array

    list_df.append(pd.DataFrame(list_newrows_dict))

clean_df = pd.concat(list_df)[right_column_order].set_index('unique_Id')
del data
```

##Set and fit the classifier

```python

# extract the data which is reasonable
is_reasonable_data = clean_df['Expected'] < 70

# use the .copy method to avoid those annoying SettingWithCopyWarning 
df = clean_df[is_reasonable_data].copy()
del clean_df

# split the data in two: whether it rained or not.
df['rain'] = df['Expected'].apply(lambda n: n > 0)


## Compute various moments of the Reflectivity data
df['Avg_Reflectivity'],  df['Range_Reflectivity'], df['Nval_Reflectivity']=\
    zip(*df['Reflectivity'].apply(clean.getListReductions))


col2fit = ['Expected', 'Avg_Reflectivity', 'Range_Reflectivity', 'Nval_Reflectivity']
values2fit = df[col2fit].values

del df #save memory



from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=30)

nfit = int(0.7*nrows)
forest.fit(values2fit[:nfit,1:], values2fit[:nfit,0])

output = forest.predict(values2fit[nfit:,1:])
#print(list(zip(values2fit[:,0], output))[:100])
```

###Simple check on how well the fit did
Eventually we should use the sklearn tools to do this

```python
score = kaggle_metric(output, values2fit[nfit:,0])

hdiff = plt.hist(N.subtract(output, values2fit[nfit:,0]), bins=100, range=(-10,10),log=True,align='left')
#hdiff = plt.hist(N.subtract(output, values2fit[nfit:,0]), bins=100,log=True)
plt.title('Predicted - True. Kaggle Score: %10.8f'%score)
```

#Zdr
Differential reflectivity in dB. Horizontal - Vertical reflectivity.

```python
## The usual header
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as N

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
import sys
sys.path.append('../../krad/')
import clean
```

```python
filename = "../Data/train_2013.csv"
nrows = 20000
data = pd.read_csv(filename, nrows=nrows)
is_reasonable_data = data['Expected'] < 70
df = data[is_reasonable_data].copy()
del data
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)
df['Zdr1'] = df[['RadarCounts','Zdr']].apply(clean.getIthRadar, axis=1)
df['Avg_Zdr'],  df['Range_Zdr'], df['Nval_Zdr']=\
    zip(*df['Zdr1'].apply(clean.getListReductions))
```

```python
ax = df['Avg_Zdr'].hist(bins=100)
ax.set_yscale('log')
_ = plt.title('All Zdr')
```

```python
fig, axes = plt.subplots(1, 2)

## Average
ax_avg = plt.subplot(1,2,1)
df['Avg_Zdr'].hist(bins=20)
plt.title('Average')
ax_avg.set_yscale('log')

## Range
ax_rge = plt.subplot(1,2,2)
df['Range_Zdr'].hist(bins=20)
plt.title('Range')
ax_rge.set_yscale('log')
```

```python
rain    = df['Expected'].apply(lambda n: n > 0 )
no_rain = df['Expected'].apply(lambda n: n == 0 )

## Avg
rain_avg_zdr = df[rain]['Avg_Zdr'].get_values()
no_rain_avg_zdr = df[no_rain]['Avg_Zdr'].get_values()

## Range
rain_range_zdr = df[rain]['Range_Zdr'].get_values()
no_rain_range_zdr = df[no_rain]['Range_Zdr'].get_values()


## Plotting
ax_rain_avg = plt.subplot(1,2,1)
avg = plt.hist([rain_avg_zdr,no_rain_avg_zdr],
               range=(-10,10),bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
ax_rain_avg.set_yscale('log')
plt.title('Average')

ax_rain_rge = plt.subplot(1,2,2)
rng = plt.hist([rain_range_zdr,no_rain_range_zdr],range=(0,20),
               bins=45,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
ax_rain_rge.set_yscale('log')
plt.title('Range')
plt.ylim([0,0.3])
```

#RhoHV
Correlation coefficient. Normalized sum of all v(hpol) - v(vpol) in the over all the radar pulses.(hpol/vpol are horizontal/vertical polarization). Should be between 0,1 but above 1 is related to noise it should not be trunckated to 1. Pure rain should be very close to 1, while many different scatteres ~0 (i.e. no correlations). Effectively, RhoHV0.97 is usually pure rain or snow

```python
import sys
sys.path.append('../../krad/')
import clean
```

```python
filename = "../Data/train_2013.csv"
nrows = 20000
data = pd.read_csv(filename, nrows=nrows)
is_reasonable_data = data['Expected'] < 70
df = data[is_reasonable_data].copy()
del data
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)
df['RhoHV1'] = df[['RadarCounts','RhoHV']].apply(clean.getIthRadar, axis=1)
df['Avg_RhoHV'],  df['Range_RhoHV'], df['Nval_RhoHV']=\
    zip(*df['RhoHV1'].apply(clean.getListReductions))
```

```python
ax = df['Avg_RhoHV'].hist(bins=100, bottom=0.1)
ax.set_yscale('log')
_ = plt.title('All RhoHV')
```

```python
fig, axes = plt.subplots(1, 2)

## Average
ax_avg = plt.subplot(1,2,1)
df['Avg_RhoHV'].hist(bins=20, bottom = 0.1)
plt.title('Average')
ax_avg.set_yscale('log')

## Range
ax_rge = plt.subplot(1,2,2)
df['Range_RhoHV'].hist(bins=20, bottom = 0.1)
plt.title('Range')
ax_rge.set_yscale('log')
```

```python
rain    = df['Expected'].apply(lambda n: n > 0 )
no_rain = df['Expected'].apply(lambda n: n == 0 )

## Avg
rain_avg = df[rain]['Avg_RhoHV'].get_values()
no_rain_avg = df[no_rain]['Avg_RhoHV'].get_values()

## Range
rain_range = df[rain]['Range_RhoHV'].get_values()
no_rain_range = df[no_rain]['Range_RhoHV'].get_values()


## Plotting
ax_rain_avg = plt.subplot(1,2,1)
avg = plt.hist([rain_avg,no_rain_avg],
               range=(0,1.5),bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
#ax_rain_avg.set_yscale('log')
plt.title('Average')
plt.ylim([0,18])

ax_rain_rge = plt.subplot(1,2,2)
rng = plt.hist([rain_range,no_rain_range],range=(0,1.5),
               bins=45,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
#ax_rain_rge.set_yscale('log')
plt.title('Range')
plt.ylim([0,10])
```

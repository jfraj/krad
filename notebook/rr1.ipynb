#RR1
Rain rate from HCA-based algorithm (HCA: Hydrometeor Classification Algorithm)
See [this web site](http://www.grlevelx.com/manuals/grlevel3_2/hca.htm) for more details.
The last plot shows a good correlation with the expected rain from the gauge.

```python
## The usual header
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as N

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
import sys
sys.path.append('../../krad/')
import clean
```

```python
filename = "../Data/train_2013.csv"
nrows = 20000
data = pd.read_csv(filename, nrows=nrows)
is_reasonable_data = data['Expected'] < 70
df = data[is_reasonable_data].copy()
del data
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)
df['RR11'] = df[['RadarCounts','RR1']].apply(clean.getIthRadar, axis=1)
df['Avg_RR1'],  df['Range_RR1'], df['Nval_RR1']=\
    zip(*df['RR11'].apply(clean.getListReductions))
```

```python
ax = df['Avg_RR1'].hist(bins=100)
ax.set_yscale('log')
_ = plt.title('All RR1')
```

```python
ax_standard = df[df['Avg_RR1']>=0]['Avg_RR1'].hist(bins=50)
ax_standard.set_yscale('log')
_ = plt.title('RR1>0')
```

```python
df['RR1ExpectedDiff'] = df['Avg_RR1'] - df['Expected']
```

```python
ax_diff = df[df['Avg_RR1']>=0]['RR1ExpectedDiff'].hist(bins=200)
ax_diff.set_yscale('log')
_ = plt.title('RR1 - Expected')
```

#RadarQualityIndex
When this feature is high (close to 1) the separation between rain and no rain is slightly clearer (see last plot)

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as N

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
import sys
sys.path.append('../../krad/')
import clean
```

```python
filename = "../Data/train_2013.csv"
nrows = 200000
data = pd.read_csv(filename, nrows=nrows)
is_reasonable_data = data['Expected'] < 70
df = data[is_reasonable_data].copy()
del data
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)
df['RadarQualityIndex1'] = df[['RadarCounts','RadarQualityIndex']].apply(clean.getIthRadar, axis=1)
df['Avg_RadarQualityIndex'],  df['Range_RadarQualityIndex'], df['Nval_RadarQualityIndex']=\
    zip(*df['RadarQualityIndex1'].apply(clean.getListReductions))
```

```python
ax = df['Avg_RadarQualityIndex'].hist(bins=100)
ax.set_yscale('log')
_ = plt.title('All RadarQualityIndex')
```

```python
qualityLower1 = df['Avg_RadarQualityIndex']<=1
qualityHigher0 = df['Avg_RadarQualityIndex']>=0
ax_standard = df[qualityHigher0 & qualityLower1]['Avg_RadarQualityIndex'].hist(bins=50)
plt.ylim([0,int(nrows/15)])
_ = plt.title('0<RadarQualityIndex<1')
```

```python
df['Reflectivity'] = df[['RadarCounts','Reflectivity']].apply(clean.getIthRadar, axis=1)
df['Avg_Reflectivity'],  df['Range_Reflectivity'], df['Nval_Reflectivity']=\
    zip(*df['Reflectivity'].apply(clean.getListReductions))
```

```python
rain    = df['Expected'].apply(lambda n: n > 2 )
no_rain = df['Expected'].apply(lambda n: n == 0 )

## Avg
rain_avg_reflectivity = df[rain]['Avg_Reflectivity'].get_values()
no_rain_avg_reflectivity = df[no_rain]['Avg_Reflectivity'].get_values()
avg = plt.hist([rain_avg_reflectivity,no_rain_avg_reflectivity],
               range=(-20,60),bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
```

Let's see if it radar quality index helps things

```python
good_radar    = df['Avg_RadarQualityIndex'] < 0.8
bad_radar    = df['Avg_RadarQualityIndex'] > 0.2



## Good radar
rain_avg_reflectivity_good = df[rain & good_radar]['Avg_Reflectivity'].get_values()
norain_avg_reflectivity_good = df[no_rain & good_radar]['Avg_Reflectivity'].get_values()

## Bad radar
rain_avg_reflectivity_bad = df[rain & bad_radar]['Avg_Reflectivity'].get_values()
norain_avg_reflectivity_bad = df[no_rain & bad_radar]['Avg_Reflectivity'].get_values()
```

```python
fig, axes = plt.subplots(3, 1)
fig.set_figheight(15)
fig.subplots_adjust(hspace=0.1)

## All radars
ax_all = plt.subplot(3,1,1)
avg_all = plt.hist([rain_avg_reflectivity,no_rain_avg_reflectivity],
               range=(-20,60),bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
ax_all.text(.2,.9,'All Radar',
        horizontalalignment='left',
        transform=ax_all.transAxes)


## Good radars
ax_good = plt.subplot(3,1,2)
avg_good = plt.hist([rain_avg_reflectivity_good,norain_avg_reflectivity_good],
               range=(-20,60),bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
ax_good.text(.2,.9,'Good Radar',
        horizontalalignment='left',
        transform=ax_good.transAxes)



## Bad radars
ax_bad = plt.subplot(3,1,3)
avg_bad = plt.hist([rain_avg_reflectivity_bad,norain_avg_reflectivity_bad],
               range=(-20,60),bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
ax_bad.text(.2,.9,'Bad Radar',
        horizontalalignment='left',
        transform=ax_bad.transAxes)
```

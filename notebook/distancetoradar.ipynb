A quick look at the distance to radar does not show any pattern for rain and no-rain.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as N

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
import sys
sys.path.append('../../krad/')
import clean
```

```python
filename = "../Data/train_2013.csv"
data = pd.read_csv(filename, nrows=200000)
```

```python
is_reasonable_data = data['Expected'] < 70
# use the .copy method to avoid those annoying SettingWithCopyWarning 
df = data[is_reasonable_data].copy()
del data
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)
```

```python
# For taking the mean of distance to radar is a waste of cpu cycles but its ok for now
df['DistanceToRadar1'] = df[['RadarCounts','DistanceToRadar']].apply(clean.getIthRadar, axis=1)
## For now, let's use the first radar only
df['Avg_DistanceToRadar'],  df['Range_DistanceToRadar'], df['Nval_DistanceToRadar']=\
    zip(*df['DistanceToRadar1'].apply(clean.getListReductions))
```

Just to make sure, let's look at the range is zero (distance to radar should not change)

```python
_ = df['Range_DistanceToRadar'].hist(bins=50)
```

##Distributions
Let's first separate the samples into few raining categories

```python
no_rain = df['Expected'].apply(lambda n: n == 0 )
verylight_rain = df['Expected'].apply(lambda n: 0 < n <= 10 )
light_rain    = df['Expected'].apply(lambda n: 10 < n <= 20 )
moderate_rain    = df['Expected'].apply(lambda n: 20 < n <= 40 )
heavy_rain    = df['Expected'].apply(lambda n: 40 < n )
##As numpy arrays
no_rain_arr = df[no_rain]['Avg_DistanceToRadar'].get_values()
verylight_rain_arr = df[verylight_rain]['Avg_DistanceToRadar'].get_values()
light_rain_arr = df[light_rain]['Avg_DistanceToRadar'].get_values()
moderate_rain_arr = df[moderate_rain]['Avg_DistanceToRadar'].get_values()
heavy_rain = df[heavy_rain]['Avg_DistanceToRadar'].get_values()
```

The plot the scaled distributions

```python
#df['Avg_DistanceToRadar'].hist(bins=50)
avg = plt.hist([no_rain_arr, verylight_rain_arr, light_rain_arr, moderate_rain_arr, heavy_rain],
               bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['No rain','Very light rain', 'Light rain', 'Moderate rain', 'Heavy rain'],alpha=0.5)
plt.legend(loc='best')
_ = plt.title('Distance to Radar')
```

No obvious structure except that it stops at 100 (must be what they mean by "scaled" in the variable definition).  There are few spikes that I do not understand (maybe some cleaning is needed).

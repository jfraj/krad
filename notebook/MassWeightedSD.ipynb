```python
## The usual header
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as N

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
import sys
sys.path.append('../../krad/')
import clean
```

```python
filename = "../Data/train_2013.csv"
nrows = 20000
data = pd.read_csv(filename, nrows=nrows)
is_reasonable_data = data['Expected'] < 70
df = data[is_reasonable_data].copy()
del data
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)
df['MassWeightedSD1'] = df[['RadarCounts','MassWeightedSD']].apply(clean.getIthRadar, axis=1)
df['Avg_MassWeightedSD'],  df['Range_MassWeightedSD'], df['Nval_MassWeightedSD']=\
    zip(*df['MassWeightedSD1'].apply(clean.getListReductions))
```

```python
ax = df['Avg_MassWeightedSD'].hist(bins=100, bottom=0.1)
ax.set_yscale('log')
_ = plt.title('All MassWeightedSD')
```

```python
fig, axes = plt.subplots(1, 2)

## Average
ax_avg = plt.subplot(1,2,1)
df['Avg_MassWeightedSD'].hist(bins=20, bottom = 0.1)
plt.title('Average')
ax_avg.set_yscale('log')

## Range
ax_rge = plt.subplot(1,2,2)
df['Range_MassWeightedSD'].hist(bins=20, bottom = 0.1)
plt.title('Range')
ax_rge.set_yscale('log')
```

```python
rain    = df['Expected'].apply(lambda n: n > 0 )
no_rain = df['Expected'].apply(lambda n: n == 0 )

## Avg
rain_avg = df[rain]['Avg_MassWeightedSD'].get_values()
no_rain_avg = df[no_rain]['Avg_MassWeightedSD'].get_values()

## Range
rain_range = df[rain]['Range_MassWeightedSD'].get_values()
no_rain_range = df[no_rain]['Range_MassWeightedSD'].get_values()


## Plotting
ax_rain_avg = plt.subplot(1,2,1)
avg = plt.hist([rain_avg,no_rain_avg],
               range=(0,3),bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
#ax_rain_avg.set_yscale('log')
plt.title('Average')
plt.ylim([0,6])

ax_rain_rge = plt.subplot(1,2,2)
rng = plt.hist([rain_range,no_rain_range],range=(0,2),
               bins=45,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
#ax_rain_rge.set_yscale('log')
plt.title('Range')
plt.ylim([0,2])
```

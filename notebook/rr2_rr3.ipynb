```python
## The usual header
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as N

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
import sys
sys.path.append('../../krad/')
import clean
```

```python
filename = "../Data/train_2013.csv"
nrows = 20000
data = pd.read_csv(filename, nrows=nrows)
is_reasonable_data = data['Expected'] < 70
df = data[is_reasonable_data].copy()
del data
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)
df['RR21'] = df[['RadarCounts','RR2']].apply(clean.getIthRadar, axis=1)
df['Avg_RR2'],  df['Range_RR2'], df['Nval_RR2']=\
    zip(*df['RR21'].apply(clean.getListReductions))

df['RR31'] = df[['RadarCounts','RR3']].apply(clean.getIthRadar, axis=1)
df['Avg_RR3'],  df['Range_RR3'], df['Nval_RR3']=\
    zip(*df['RR31'].apply(clean.getListReductions))
```

```python
plt.subplot(2,1,1)
ax = df['Avg_RR2'].hist(bins=100)
ax.set_yscale('log')
_ = plt.title('All RR2')
plt.subplot(2,1,2)
ax = df['Avg_RR3'].hist(bins=100)
ax.set_yscale('log')
_ = plt.title('All RR3')
```

```python
plt.subplot(2,1,1)
ax_standard = df[df['Avg_RR2']>=0]['Avg_RR2'].hist(bins=50)
ax_standard.set_yscale('log')
_ = plt.title('RR2>0')
plt.subplot(2,1,2)
ax_standard = df[df['Avg_RR3']>=0]['Avg_RR3'].hist(bins=50)
ax_standard.set_yscale('log')
_ = plt.title('RR3>0')
```

```python
df['RR2ExpectedDiff'] = df['Avg_RR2'] - df['Expected']
df['RR3ExpectedDiff'] = df['Avg_RR3'] - df['Expected']
```

```python
plt.subplot(2,1,1)
ax_diff = df[df['Avg_RR2']>=0]['RR2ExpectedDiff'].hist(bins=200,range=(-20,30))
ax_diff.set_yscale('log')
_ = plt.title('RR2 - Expected')
plt.subplot(2,1,2)
ax_diff = df[df['Avg_RR3']>=0]['RR3ExpectedDiff'].hist(bins=200,range=(-20,30))
ax_diff.set_yscale('log')
_ = plt.title('RR3 - Expected')
```

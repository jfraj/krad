#Study of Reflectivity and Range
This study shows that the range of reflectivity is a potentially useful feature to preditc the amount of rain.
Overview:
* Clean the data set
* Creates column with average, range (max - min) and number of measurements of reflectivity
* Look at the combination of the range and average as an improved rain indicator
* A little look at the range normalized by the average reflectivity (not conclusive)

Note that I use single radars only.

```python
# The usual preamble, following Jula Evans' tutorial 
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as N

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
import sys
sys.path.append('../../krad/')
import clean
```

```python
filename = "../Data/train_2013.csv"
data = pd.read_csv(filename, nrows=200000)
```

##Some cleaning

```python
is_reasonable_data = data['Expected'] < 70
df = data.loc[is_reasonable_data]
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)

# Get the single radar data
def identify_single_radar(radar_tuple):
    #if radar_tuple[1] == 0:
    if len(radar_tuple) < 2:
        return True
    else:
        return False
    
##For now lets look at only one-radar rows
single_radar = df['RadarCounts'].apply(identify_single_radar)
df_single_radar = df.loc[single_radar]
df_single_radar['Avg_Reflectivity'],  df_single_radar['Range_Reflectivity'], df_single_radar['Nval_Reflectivity']=\
    zip(*df_single_radar['Reflectivity'].apply(clean.getListReductions))

```

##Turn the reflectivity list into Avg, rang, reflectivity columns

```python
df_single_radar[['Reflectivity', 'Avg_Reflectivity', 'Range_Reflectivity', 'Nval_Reflectivity']].head()
```

###Let's plot these three variables separately

```python
fig, axes = plt.subplots(1, 3)

## Average
plt.subplot(1,3,1)
df_single_radar['Avg_Reflectivity'].hist(bins=20)
plt.title('Average')

## Range
plt.subplot(1,3,2)
df_single_radar['Range_Reflectivity'].hist(bins=20)
plt.title('Range')

## Number of values
plt.subplot(1,3,3)
df_single_radar['Nval_Reflectivity'].hist(bins=20)
plt.title('# of values')
```

##Effect on rain

```python
rain    = df_single_radar['Expected'].apply(lambda n: n > 0 )
no_rain = df_single_radar['Expected'].apply(lambda n: n == 0 )

## Avg
rain_avg_reflectivity = df_single_radar[rain]['Avg_Reflectivity'].get_values()
no_rain_avg_reflectivity = df_single_radar[no_rain]['Avg_Reflectivity'].get_values()

## Range
rain_range_reflectivity = df_single_radar[rain]['Range_Reflectivity'].get_values()
no_rain_range_reflectivity = df_single_radar[no_rain]['Range_Reflectivity'].get_values()

## N values
rain_nval_reflectivity = df_single_radar[rain]['Nval_Reflectivity'].get_values()
no_rain_nval_reflectivity = df_single_radar[no_rain]['Nval_Reflectivity'].get_values()



## Plotting
plt.subplot(1,3,1)
avg = plt.hist([rain_avg_reflectivity,no_rain_avg_reflectivity],bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
plt.title('Average')

plt.subplot(1,3,2)
rng = plt.hist([rain_range_reflectivity,no_rain_range_reflectivity],bins=45,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
plt.title('Range')
plt.ylim([0,0.08])

plt.subplot(1,3,3)
nval = plt.hist([rain_nval_reflectivity,no_rain_nval_reflectivity],bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
plt.title('# of values')
```

As expected the average is the most important parameter, and the number of values does not seem to help.  
Rain data seems to have a slightly higher reflectivity range than no-rain data.
This might because of variable raining conditions: i.e. it might rain only in the first 30 minutes.
Let see if it helps by when the range is combined with the average.

```python
ax = df_single_radar[no_rain].plot(kind='scatter', x='Avg_Reflectivity', y='Range_Reflectivity', color='Red', label='No Rain')
df_single_radar[rain].plot(kind='scatter', x='Avg_Reflectivity', y='Range_Reflectivity', color='Blue', label='Rain', ax=ax)
```

It might help to see the no-rain data separately

```python
df_single_radar[no_rain].plot(kind='scatter', x='Avg_Reflectivity', y='Range_Reflectivity', color='Red', label='No Rain')
```

Or as 2D histograms

```python
from matplotlib.colors import LogNorm
fig2d = plt.figure(figsize=(12, 12))
fig2d.subplots_adjust(hspace=.01)
rainsubplt = plt.subplot(2,1,1)
rng2D = plt.hist2d(rain_avg_reflectivity, rain_range_reflectivity, bins=50, norm=LogNorm(), range=[[-20, 40], [0, 50]])
#plt.title('Rain')
plt.ylabel('Range')
rainsubplt.text(27, 46, 'Rain', fontsize=35, color='Blue')

norainsubplt = plt.subplot(2,1,2)
rng2D = plt.hist2d(no_rain_avg_reflectivity, no_rain_range_reflectivity, bins=50, norm=LogNorm(), range=[[-20, 40], [0, 50]])
#plt.title('No rain')
plt.ylabel('Range')
plt.xlabel('Average')
norainsubplt.text(27, 46, 'No Rain', fontsize=35, color='Red')
```

There is an obvious difference between rain and no rain and not all comes from the average reflectivity.  For example, it seems that the combination of high reflectivity and high range (i.e. large reflectivity variation) favors rains whereas high reflectivity and low range favors no rain.  It will be up to the classifier to decide the optimal selection.

##Not so useful normalized distributions
One would think that the higher the reflection, the higher the range it will have.  Although this can be taken care by the classifier (I assume), let's see if we can see something interesting with the range normalized by the reflectivity.  Let's first create a new column Normrange_Reflectivity = Range_Reflectivity/Avg_Reflectivity :

```python
df_single_radar['Normrange_Reflectivity'] = df_single_radar['Range_Reflectivity']/(df_single_radar['Avg_Reflectivity'].abs() + 0.1)
df_single_radar[['Avg_Reflectivity','Range_Reflectivity', 'Normrange_Reflectivity']].head()
```

```python
rain_nrange_reflectivity = df_single_radar[rain]['Normrange_Reflectivity'].get_values()
no_rain_nrange_reflectivity = df_single_radar[no_rain]['Normrange_Reflectivity'].get_values()

fignrng = plt.figure()
#nrng = plt.hist([rain_nrange_reflectivity,no_rain_nrange_reflectivity],bins=10000,normed=True,stacked=False,histtype='stepfilled',
#                label=['Rain','No Rain'],alpha=0.75)
nrng = plt.hist([rain_nrange_reflectivity,no_rain_nrange_reflectivity],bins=10000,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.legend(loc='best')
plt.yscale('log')
#plt.xscale('log')
plt.title('Normalized Range')
plt.xlim([0,15])
#plt.xlim([0,30])
```

```python
fignorm = plt.figure()
#nax = df_single_radar[no_rain].plot(kind='scatter', x='Avg_Reflectivity', y='Normrange_Reflectivity', color='Red', label='No Rain', logy=True, logx=True)
nax = df_single_radar[no_rain].plot(kind='scatter', x='Avg_Reflectivity', y='Normrange_Reflectivity', color='Red', label='No Rain')
df_single_radar[rain].plot(kind='scatter', x='Avg_Reflectivity', y='Normrange_Reflectivity', color='Blue', label='Rain', ax=nax)
plt.ylim([0.01,5])
plt.xlim([-20,50])
```

```python
fig2dn = plt.figure(figsize=(12, 12))
fig2dn.subplots_adjust(hspace=.01)
nrainsubplt = plt.subplot(2,1,1)
nrng2D = plt.hist2d(rain_avg_reflectivity, rain_nrange_reflectivity, bins=150, norm=LogNorm(), range=[[-20, 50], [0.01, 5]])
#plt.yscale('log')
plt.ylabel('Normalized Range')
#nrainsubplt.text(32, 80, 'Rain', fontsize=50, color='Blue')
plt.ylim([0.01,5])
plt.xlim([-20,50])

nnorainsubplt = plt.subplot(2,1,2)
nrng2D = plt.hist2d(no_rain_avg_reflectivity, no_rain_nrange_reflectivity, bins=150, norm=LogNorm(), range=[[-20, 50], [0.01, 5]])
#plt.yscale('log')
plt.ylabel('Normalized Range')
plt.xlabel('Average')
#nnorainsubplt.text(32, 80, 'No Rain', fontsize=50, color='Red')
plt.ylim([0.01,5])
plt.xlim([-20,50])
```

Similarly with the reflectivity range, its normalized version improves the separation 
but it is not clear whether it makes a real improvement and the classifier might be better to decide.

# In this notebook, I will explore some of the features of the Doppler radar data for the rain prediction Kaggle competition.

Column explanations

- TimeToEnd:  How many minutes before the end of the hour was this radar observation?

- DistanceToRadar:  Distance between radar and gauge.  This value is scaled and rounded to prevent reverse engineering gauge location

- Composite:  Maximum reflectivity in vertical volume above gauge

- HybridScan: Reflectivity in elevation scan closest to ground

- HydrometeorType:  One of nine categories in NSSL HCA. See presentation for details.

- Kdp:  Differential phase

- RR1:  Rain rate from HCA-based algorithm

- RR2:  Rain rate from Zdr-based algorithm

- RR3:  Rain rate from Kdp-based algorithm

- RadarQualityIndex:  A value from 0 (bad data) to 1 (good data)

- Reflectivity:  In dBZ

- ReflectivityQC:  Quality-controlled reflectivity

- RhoHV:  Correlation coefficient

- Velocity:  (aliased) Doppler velocity

- Zdr:  Differential reflectivity in dB

- LogWaterVolume:  How much of radar pixel is filled with water droplets?

- MassWeightedMean:  Mean drop size in mm

- MassWeightedSD:  Standard deviation of drop size

- Expected: the actual amount of rain reported by the rain gauge for that hour (in mm)

```python
import sys
sys.path.append('../../krad/')
import clean


# remove unreasonable gauge data
is_reasonable_data = data['Expected'] < 70

# let's create a copy of the dataframe with only the reasonable data.
# the use of the method .copy creates a completely new dataframe which
# is no longer tied with the original one (this isn't very efficient as we are
# copying data). However, this avoids those annoying SettingWithCopyWarning.
df = data[is_reasonable_data].copy()


# Create a new column which contains a tuple representing
# the number of distinct radar measurements in the row, each
# element of the tuple indicating the number of time steps for a given
# radar
df['RadarCounts'] = df['TimeToEnd'].apply(clean.getRadarLength)

    # Get the single radar data
def identify_single_radar(radar_tuple):
    if len(radar_tuple) == 1:
        return True
    else:
        return False
    
single_radar = df['RadarCounts'].apply(identify_single_radar)


# let's use the .copy() method again, avoiding SettingWithCopyWarning.
df_single_radar = df[single_radar].copy()


# split the data in two: whether it rained or not.
rain    = df_single_radar['Expected'].apply(lambda n: n > 0 )
no_rain = df_single_radar['Expected'].apply(lambda n: n == 0 )

df_single_radar['Avg_Reflectivity'] = (df_single_radar[['RadarCounts','Reflectivity']]
                                       .apply(clean.separate_listInColumn,axis=1)
                                       .apply(lambda tup: N.average(tup[0])))

rain_avg_reflectivity = df_single_radar[rain]['Avg_Reflectivity'].get_values()
no_rain_avg_reflectivity = df_single_radar[no_rain]['Avg_Reflectivity'].get_values()

hist = plt.hist([rain_avg_reflectivity,no_rain_avg_reflectivity],bins=100,normed=True,stacked=False,histtype='stepfilled',
                label=['Rain','No Rain'],alpha=0.75)
plt.xlabel('Average Reflectivity')
plt.ylabel('Probability density')
plt.title('Distribution of average reflectivity, depending on whether it rained or not')
plt.legend(loc=0)
```

```python
#It's not easy to differentiate between the 2. WHat if we segment with more categories? Like light rain vs heavy rain?
#According to Wikipedia (http://en.wikipedia.org/wiki/Rain)
#Light rain : <2.5 mm / hour
#Moderate rain : <10 mm / hour
#Heavy rain : <50mm / hour
#Violent rain : >50mm / hour

no_rain       = df_single_radar_valid['Expected'].apply(lambda n: n == 0 )
light_rain    = df_single_radar_valid['Expected'].apply(lambda n: (n > 0) and (n<=2.5))
moderate_rain = df_single_radar_valid['Expected'].apply(lambda n: (n > 2.5) and (n<=10.))
heavy_rain    = df_single_radar_valid['Expected'].apply(lambda n: (n > 10))

no_rain_avg_reflectivity       = df_single_radar_valid[no_rain]['MeanReflectivityQC'].get_values()
light_rain_avg_reflectivity    = df_single_radar_valid[light_rain]['MeanReflectivityQC'].get_values()
moderate_rain_avg_reflectivity = df_single_radar_valid[moderate_rain]['MeanReflectivityQC'].get_values()
heavy_rain_avg_reflectivity    = df_single_radar_valid[heavy_rain]['MeanReflectivityQC'].get_values()

hist = plt.hist([no_rain_avg_reflectivity, light_rain_avg_reflectivity, moderate_rain_avg_reflectivity,
                 heavy_rain_avg_reflectivity],
                bins=100,
                normed=True,
                stacked=False,
                histtype='stepfilled',
                label=['No Rain', 'Light', 'Moderate', 'Heavy'],
                alpha=0.75)
plt.xlabel('Average Reflectivity')
plt.ylabel('Probability density')
plt.title('Distribution of average reflectivity, depending on whether it rained or not')
plt.legend(loc=0)
```

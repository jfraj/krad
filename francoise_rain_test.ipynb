```python

# remove unreasonable gauge data
total_samples= len(data)
is_reasonable_data = data['Expected'] <= 70
df = data.loc[is_reasonable_data]
df.head(2)
is_valid = df['ReflectivityQC'].apply(lambda s :'-9990' not in s)
valid_sample = is_valid.sum()
print "Proportion of valid samples in dataset is : {}%".format(float(valid_sample)/total_samples*100.)
```

```python
df_valid = df.loc[is_valid]
```

```python
df_valid.head()
```

```python
df_valid['MeanReflectivityQC'] = df_valid['ReflectivityQC'].apply(lambda s : np.array(map(float, s.split())).mean())
```

```python
df_valid.head(2)
```

```python
curated_data = df_valid[['MeanReflectivityQC', 'Expected']].values
curated_data
```

```python
#It's not easy to differentiate between the 2. WHat if we segment with more categories? Like light rain vs heavy rain?
#According to Wikipedia (http://en.wikipedia.org/wiki/Rain)
#Light rain : <2.5 mm / hour
#Moderate rain : <10 mm / hour
#Heavy rain : <50mm / hour
#Violent rain : >50mm / hour

no_rain       = df_valid['Expected'].apply(lambda n: n == 0 )
light_rain    = df_valid['Expected'].apply(lambda n: (n > 0) and (n<=2.5))
moderate_rain = df_valid['Expected'].apply(lambda n: (n > 2.5) and (n<=10.))
heavy_rain    = df_valid['Expected'].apply(lambda n: (n > 10))

no_rain_avg_reflectivity       = df_valid[no_rain]['MeanReflectivityQC'].get_values()
light_rain_avg_reflectivity    = df_valid[light_rain]['MeanReflectivityQC'].get_values()
moderate_rain_avg_reflectivity = df_valid[moderate_rain]['MeanReflectivityQC'].get_values()
heavy_rain_avg_reflectivity    = df_valid[heavy_rain]['MeanReflectivityQC'].get_values()

hist = plt.hist([no_rain_avg_reflectivity, light_rain_avg_reflectivity, moderate_rain_avg_reflectivity,
                 heavy_rain_avg_reflectivity],
                bins=100,
                normed=True,
                stacked=False,
                histtype='stepfilled',
                label=['No Rain', 'Light', 'Moderate', 'Heavy'],
                alpha=0.75)
plt.xlabel('Average Reflectivity')
plt.ylabel('Probability density')
plt.title('Distribution of average reflectivity, depending on whether it rained or not')
plt.legend(loc=0)
```

```python
df_valid['StdReflectivityQC'] = df_valid['ReflectivityQC'].apply(lambda s : np.array(map(float, s.split())).std())
```

```python
no_rain_std_reflectivity       = df_valid[no_rain]['StdReflectivityQC'].get_values()
light_rain_std_reflectivity    = df_valid[light_rain]['StdReflectivityQC'].get_values()
moderate_rain_std_reflectivity = df_valid[moderate_rain]['StdReflectivityQC'].get_values()
heavy_rain_std_reflectivity    = df_valid[heavy_rain]['StdReflectivityQC'].get_values()

hist = plt.hist([no_rain_std_reflectivity, light_rain_std_reflectivity, moderate_rain_std_reflectivity,
                 heavy_rain_std_reflectivity],
                bins=100,
                normed=True,
                stacked=False,
                histtype='stepfilled',
                label=['No Rain', 'Light', 'Moderate', 'Heavy'],
                alpha=0.75)
plt.xlabel('Standard deviation of Reflectivity')
plt.ylabel('Probability density')
plt.title('Distribution of std reflectivity, depending on whether it rained or not')
plt.legend(loc=0)
```

```python
df_valid['StdNorm'] = df_valid['StdReflectivityQC']/df_valid['MeanReflectivityQC']
```

```python
no_rain_stdn_reflectivity       = df_valid[no_rain]['StdNorm'].get_values()
light_rain_stdn_reflectivity    = df_valid[light_rain]['StdNorm'].get_values()
moderate_rain_stdn_reflectivity = df_valid[moderate_rain]['StdNorm'].get_values()
heavy_rain_stdn_reflectivity    = df_valid[heavy_rain]['StdNorm'].get_values()

hist = plt.hist([no_rain_stdn_reflectivity, light_rain_stdn_reflectivity, moderate_rain_stdn_reflectivity,
                 heavy_rain_stdn_reflectivity],
                bins=100,
                normed=True,
                stacked=False,
                histtype='stepfilled',
                label=['No Rain', 'Light', 'Moderate', 'Heavy'],
                alpha=0.75)
plt.xlabel('Standard deviation of Reflectivity normalized')
plt.ylabel('Probability density')
plt.title('Distribution of std reflectivity, depending on whether it rained or not')
plt.legend(loc=0)
```

So it's about the same standard deviation distribution, no matter what the amount of rain.
